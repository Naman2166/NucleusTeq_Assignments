```python
# importing libraries 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# Creating student dataset

students = {
    "Name": ["Naman", "Ronaldo", "Messi", "Virat"],
    "Marks": [70, 80, 90, 60],
    "Hours Studied": [2, 3, 5, 1]
}

students
```

```python
# 1. Load student dataset into Pandas

students_dataframe = pd.DataFrame(students)

students_dataframe
```

```python
# 2. Add Performance column

students_dataframe["Performance"] = [ 
    ("Pass" if marks > 65 else "Fail") for marks in students_dataframe["Marks"]
]

students_dataframe
```

```python
# 3. Create line chart: Hours vs Marks

plt.figure(figsize=(6, 4))
plt.plot( students_dataframe["Hours Studied"], students_dataframe["Marks"], marker="o")
plt.title("Hours Studied vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.show()
```

```python
# 4. Create scatter plot: Hours Studied vs Marks

plt.figure(figsize=(6, 4))
plt.scatter(students_dataframe["Hours Studied"], students_dataframe["Marks"])
plt.title("Hours Studied vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.show()
```

```python
# 5. Create seaborn barplot: Performance vs Marks

plt.figure(figsize=(6, 4))
sb.barplot(data=students_dataframe, x="Performance", y="Marks")
plt.title("Performance vs Marks")
plt.show()
```
