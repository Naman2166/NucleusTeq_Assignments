```python
# importing pandas library 

import pandas as pd
```

```python
# Creating Employee DataFrame

employee_dataframe = pd.DataFrame({
    "Name": ["Naman", "Ronaldo", "Messi", "Virat"],
    "Age": [23, 30, 28, 35],
    "Department": ["IT", "HR", "Finance", "IT"],
    "Salary": [30000, 50000, 45000, 60000]
})

employee_dataframe
```
