```python
# importing pandas library

import pandas as pd
```

```python
# creating dataset with missing values

employee_dataframe = pd.DataFrame({
    "Name": ["Rahul", "Priya", "Anuj"],
    "Age": [25, None, 29],
    "Salary": [30000, 40000, None]
})

employee_dataframe
```

```python
# 1. Detect missing values

employee_dataframe.isnull()
```

```python
# 2. Replace missing Age with mean

# calculating mean
mean_age = employee_dataframe["Age"].mean()
print("Mean age:", mean_age)

# replacing missing age with mean_age
employee_dataframe["Age"] = employee_dataframe["Age"].fillna(mean_age)

employee_dataframe
```

```python
# 3. Replace missing Salary with 0

employee_dataframe["Salary"] = employee_dataframe["Salary"].fillna(0)

employee_dataframe
```
