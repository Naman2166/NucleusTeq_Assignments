```python
# importing pandas library 

import pandas as pd
```

```python
# Creating Employee DataFrame

employee_dataframe = pd.DataFrame({
    "Name": ["Naman", "Ronaldo", "Messi", "Virat", "Rohit"],
    "Age": [23, 30, 28, 29, 35],
    "Department": ["IT", "HR", "Finance", "Finance", "IT"],
    "Salary": [30000, 50000, 45000, 47000, 60000]
})

employee_dataframe
```

```python
# 1. Find average salary by department

departments = employee_dataframe.groupby("Department")

departments["Salary"].mean()
```

```python
# 2. Find max salary by department

departments = employee_dataframe.groupby("Department")

departments["Salary"].max()
```

```python
# 3. Count employees per department

departments = employee_dataframe.groupby("Department")

departments["Name"].count()
```
