```python
# importing libraries 

import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# Creating Employee DataFrame

employee_dataframe = pd.DataFrame({
    "Name": ["Naman", "Ronaldo", "Messi", "Virat", "Rohit"],
    "Age": [23, 30, 28, 29, 35],
    "Department": ["IT", "HR", "Finance", "Finance", "IT"],
    "Salary": [42000, 50000, 45000, 47000, 60000]
})

employee_dataframe
```

```python
# 1. Create barplot: Department vs Salary

plt.figure(figsize=(6, 4))
sb.barplot(data=employee_dataframe, x="Department", y="Salary")
plt.title("Department vs Salary")
plt.show()
```

```python
# 2. Create boxplot: Salary Distribution

plt.figure(figsize=(6, 4))
sb.boxplot(y=employee_dataframe["Salary"])
plt.title("Salary Distribution")
plt.show()
```

```python
# 3. Create heatmap: Correlation between Age & Salary

correlation = employee_dataframe[["Age", "Salary"]].corr()

plt.figure(figsize=(5, 4))
sb.heatmap(correlation, annot=True)
plt.title("Age and Salary Correlation")
plt.show()
```
